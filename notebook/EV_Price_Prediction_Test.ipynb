```python
## Step 2: Load and Clean the Data
import pandas as pd

train_data = pd.read_csv('dataset/train.csv')
train_data = train_data.replace(to_replace ="N/", value=0)
test_data = pd.read_csv('dataset/test.csv')
cols = ['ID', 'VIN (1-10)', 'State', 'Electric Range', 'Base MSRP', 'DOL Vehicle ID', 'Vehicle Location', 'Electric Utility']
train_data
```

```python
## Step 3: Separate Features and Labels

from sklearn.preprocessing import OneHotEncoder as OHEncoder

X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]

X_train = X_train.drop(columns=cols)
```

```python
## Step 5: One-Hot Encode Categorical Data

Encoder_train = X_train

temp_df = pd.DataFrame({'County': counties_e})
Encoder_train = pd.concat([Encoder_train, temp_df])
temp_df = pd.DataFrame({'City': cities_e})
Encoder_train = pd.concat([Encoder_train, temp_df])
temp_df = pd.DataFrame({'ZIP Code': zips_e})
Encoder_train = pd.concat([Encoder_train, temp_df])


encoder = OHEncoder()
encoder.fit(Encoder_train)
X_train = encoder.transform(X_train)
```

```python
## Step 6: Train the Model

from sklearn.ensemble import RandomForestRegressor as RFR
rfr = RFR()
rfr.fit(X_train, y_train)
```

```python
## Got from ChatGPT. I changed "model" to "rfr"
# In your notebook, after training the model
import joblib

# Save the trained model
joblib.dump(rfr, 'model/ev_price_model.pkl')

# Save the encoder if you're using OneHotEncoder
joblib.dump(encoder, 'model/encoder.pkl')
```

```python
## Step 7: Prepare Test Data and Make Predictions

test_data_ID = test_data.ID
X_test = test_data.drop(columns=cols)

X_test = encoder.transform(X_test)
```

```python
y_pred = rfr.predict(X_test)
```

```python
## Step 8: Save Predictions to a CSV File

res_df = pd.DataFrame({
    'ID': test_data_ID,
    'Expected Price ($1k)': y_pred
})
res_df.to_csv('result-4.csv', index=False)
```
